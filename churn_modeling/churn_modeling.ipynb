```python
# reading the file
data = pd.read_csv('Churn_Modelling (1).csv')

data.head()
```

```python
data.info()
```

```python
# split X and Y

X = data.iloc[: , 3:13]
y = data.iloc[: , 13]
```

```python
X
```

```python
y
```

```python
geography = pd.get_dummies(X['Geography'] , drop_first=True)
gender = pd.get_dummies(X['Gender'] , drop_first=True)
```

```python
X = pd.concat([X , geography, gender ] , axis = 1)
```

```python
X
```

```python
X = X.drop(columns = ['Geography' , 'Gender'] , axis = 1)
```

```python
X
```

```python
# Spillting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2 , random_state = 101)
```

```python
# feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
#1. architecture creation
#2. compile
#3 . model train
```

```python
#1. model creation

model = Sequential()

model.add(Dense(200, activation='relu' , kernel_initializer='he_uniform')) # hidden layer 1
model.add(Dropout(0.2))

model.add(Dense(50 , activation='sigmoid')) # hidden layer 2
model.add(Dropout(0.5))

model.add(Dense(25 , activation='relu' , kernel_initializer='he_normal')) # hidden layer 3


model.add(Dense(1 , activation='sigmoid')) # output layer
```

```python
model.compile(optimizer='SGD' , loss = 'binary_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(X_train , y_train , epochs = 70 , batch_size=32)
```

```python
# testing the data
model.evaluate(X_test , y_test )
```
